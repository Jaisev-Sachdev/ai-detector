## Single Image Predictor Script

```python
import tensorflow as tf
import numpy as np
import os
from tensorflow.keras.preprocessing import image

# ==========================================
# CONFIGURATION
# ==========================================
# Use explicit Downloads folder so '__file__' isn't required in notebooks
BASE_DIR = r"C:\Users\jaise\Downloads"
MODEL_PATH = os.path.join(BASE_DIR, "dataset", "ai_detector_model.keras")


def load_and_predict(image_path):
    # 1. Load the model
    if not os.path.exists(MODEL_PATH):
        print(f"Error: Model not found at {MODEL_PATH}")
        return

    print("Loading model...")
    model = tf.keras.models.load_model(MODEL_PATH)

    # 2. Load and Preprocess the Image
    if not os.path.exists(image_path):
        print(f"Error: Image not found at {image_path}")
        return

    img = image.load_img(image_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Make a batch of 1
    img_array = img_array.astype('float32') / 255.0  # Normalize pixel values (0-1)

    # 3. Predict
    print("Analyzing image...")
    prediction = model.predict(img_array)
    score = float(prediction.flatten()[0])

    # 4. Output Results
    print("\n" + "=" * 30)
    print(f"RESULTS FOR: {os.path.basename(image_path)}")
    print("=" * 30)

    if score > 0.5:
        print("VERDICT:  ðŸ¤– AI GENERATED")
        confidence = score * 100
    else:
        print("VERDICT:  ðŸ“¸ REAL PHOTO")
        confidence = (1 - score) * 100

    print(f"Confidence: {confidence:.2f}%")
    print(f"Raw Score:  {score:.4f}")
    print("=" * 30 + "\n")


# Non-blocking default run for notebooks
if __name__ == "__main__":
    default_image = r"C:\Users\jaise\Downloads\unnamed.jpg"
    if os.path.exists(default_image):
        load_and_predict(default_image)
    else:
        print(f"Default image not found at {default_image}.")
        print("To test a different image, call load_and_predict(<path>) from another cell.")
```
